```r
#for the integrum federal corpus
#load documents
prep1 <- readRDS('prep1_integrum_federal.rds')
```

```r
#insert topic numbers and name the picture you are generating
# i.e. topics = c(HERE NUMBERS OF TOPICS)
#   main = "HERE LABLE YOUR PIC"
# do not change the pther parameters
plot(
  prep1, "year", 
  method = "continuous", 
  topics = c(3,12,20), 
  printlegend = TRUE,
  xaxt = "n",
  xlab = "Years",
  main = "TopicCultFed")
axis(1,at=seq(from=1997, to=2017,by=1))
```
